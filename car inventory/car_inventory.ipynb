```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import data 
data = df = pd.read_excel ('cardatanew.xlsx')
```

```python
data
```

```python
# fill make columns depend on car id column 
data['Make']=data['Make'].fillna(data['Car ID'].str[0:2])
data.head()
```

```python
data['Make (Full Name)']=data['Make (Full Name)'].fillna(data['Car ID'].str[0:2])

data.head()
```

```python
data['Make (Full Name)'].mask(data['Make (Full Name)'] == 'FD', 'Ford', inplace=True)
data['Make (Full Name)'].mask(data['Make (Full Name)'] == 'CR', 'Chrysler', inplace=True)
data['Make (Full Name)'].mask(data['Make (Full Name)'] == 'GM','General Motors', inplace=True)
data['Make (Full Name)'].mask(data['Make (Full Name)'] == 'HO','Honda', inplace=True)
data['Make (Full Name)'].mask(data['Make (Full Name)'] == 'TY','Hundal', inplace=True)
data['Make (Full Name)'].mask(data['Make (Full Name)'] == 'TO','Toyota', inplace=True)

```

```python
def make_full():
    for row in data['Make (Full Name)']:
        if row =="FD":
            row='Ford'
        elif row=='CR':
            row='Chrysler'
        elif row=='GM':
            row='General Motors'
        elif row=='HO':
            row='Honda'
        elif row=='HY':
            row='Hundal'
        else:
            row='Toyota'
make_full()
```

```python
data.head()
```

```python
# fill model columns depend on car id column 
data['Model']=data['Model'].fillna(data['Car ID'].str[4:7])
data.head()
```

```python
data['Model'].value_counts()
```

```python
data['Model (Full Name)']= data['Model']

data['Model (Full Name)'].mask(data['Model (Full Name)'] == 'CAM', 'Camrey', inplace=True)
data['Model (Full Name)'].mask(data['Model (Full Name)'] == 'CAR', 'Caravan', inplace=True)
data['Model (Full Name)'].mask(data['Model (Full Name)'] == 'CIV', 'Civic', inplace=True)
data['Model (Full Name)'].mask(data['Model (Full Name)'] == 'CMR', 'Camero', inplace=True)
data['Model (Full Name)'].mask(data['Model (Full Name)'] == 'COR', 'Corola', inplace=True)
data['Model (Full Name)'].mask(data['Model (Full Name)'] == 'ELA', 'Focus', inplace=True)
data['Model (Full Name)'].mask(data['Model (Full Name)'] == 'FCS', 'Focus', inplace=True)
data['Model (Full Name)'].mask(data['Model (Full Name)'] == 'MTG', 'Mustang', inplace=True)
data['Model (Full Name)'].mask(data['Model (Full Name)'] == 'ODY', 'Odyssey', inplace=True)
data['Model (Full Name)'].mask(data['Model (Full Name)'] == 'PTC', 'PT Cruiser', inplace=True)
data['Model (Full Name)'].mask(data['Model (Full Name)'] == 'SLV', 'Silverado', inplace=True)

```

```python
data.head()
```

```python
##fil  Manufacture Year column 
data['Manufacture Year']=data['Manufacture Year'].fillna((data['Car ID'].str[2:4]))
```

```python
data.head()
```

```python
data['Manufacture Year'].value_counts()
```

```python
# in above we have invalid value so we must processing it 
data['Manufacture Year'].mask(data['Manufacture Year'] == 'O6','06', inplace=True)
data['Manufacture Year'].mask(data['Manufacture Year'] == 'O5','05', inplace=True)
data['Manufacture Year'].mask(data['Manufacture Year'] == 'O9','09', inplace=True)

data['Manufacture Year']=data['Manufacture Year'].astype(int)

def full_year(y):
    if y <= 22:
        #y=str(y)
        y=2000+y
        return y
    else:
        #y=str(y)
        y=1900+y
        return y
    
data['Manufacture Year']=data['Manufacture Year'].apply(lambda x:full_year(x))
```

```python
data.head()
```

```python
# find age from year 
#data['Age']=data['Manufacture Year'].astype(int)
```

```python
## fill age column 

data['Age'].fillna(2022-data['Manufacture Year'],inplace=True)
data.head()
```

```python
## fill Mile/Year column
data['Miles / Year'].fillna(data['Miles']/data['Age'],inplace=True)
data.head()
```

```python
# find covered column
data['Covered?']=data['Covered?'][data['Miles']<data['Warantee Miles']].fillna('yes')
data.head()
```

```python
data['Covered?'].fillna('No',inplace=True)
data.head(100)
```

```python
# we need find new car id column so we should add color to id 
data['New Car ID']=data['Make']+data['Car ID'].str[2:4]+data['Model']+data['Color'].str[0:3].str.upper()+data['Car ID'].str[7:]
```
